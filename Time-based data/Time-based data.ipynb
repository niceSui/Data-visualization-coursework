```python
import pandas as pd
import seaborn as sns
import matplotlib
matplotlib.use('module://ipympl.backend_nbagg')
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import mplcursors

url = 'https://raw.githubusercontent.com/HKUST-VISLab/coding-challenge/master/temperature_daily.csv'
df = pd.read_csv(url)
df[['year', 'month', 'day']] = df['date'].str.split('-', expand=True)

df[['year', 'month', 'day']] = df[['year', 'month', 'day']].astype(int)
df = df[df['year']>=2008]
df_GroupByYear = df.groupby('year')[['max_temperature','min_temperature']].agg('mean')

sns.set_style("whitegrid", {'grid.color': '#EEEEEE', 'grid.alpha': 0.01})

plt.figure(figsize=(12, 6))
fig, ax = plt.subplots(figsize=(10, 4))
# sns.scatterplot(data=df_GroupByYear, x='year', y='max_temperature', color='blue')
# sns.scatterplot(data=df_GroupByYear, x='year', y='min_temperature', color='red')
sns.lineplot(data=df_GroupByYear, x='year', y='max_temperature', label='Avg Max Temperature')
sns.lineplot(data=df_GroupByYear, x='year', y='min_temperature', label='Avg Min Temperature')


ax.set_xlabel('Year')
ax.set_ylabel('Temperature')
plt.title('Average Max and Min Temperature by Year')
ax.set_ylim(bottom=0, top=40)  
fmt = '{x:.0f}C' # define the format string for y-axis labels
tick = mtick.StrMethodFormatter(fmt)
ax.xaxis.set_major_locator(mtick.MultipleLocator(1))
ax.yaxis.set_major_formatter(tick)
plt.legend()

    
plt.show()
```
